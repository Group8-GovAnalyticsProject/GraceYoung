```python
import pandas as pd

link="2019-4-cleaned.csv"

df1=pd.read_csv(link)
```

```python
#Check the contents of this table.
df1
```

```python
#Drop the rows with missing values.
df1.dropna(inplace = True)
```

```python
df1
```

```python
#See the type of each variable.
df1.dtypes
```

```python
#Change the type of the 'cbgs_id' variable to string for futher cleanup.
df1['cbgs_id'] = df1['cbgs_id'].astype('string')
```

```python
#The type of the 'count' variable is supposed to be integer. Changing the type.
df1['count'] = df1['count'].astype('int64')
```

```python
df1
```

```python
#The last 3 characters of'cbgs_id' need to be removed to get the tract id. Dropping the characters for each entry.
drop_last3 = lambda s:s[:-3]
df1.loc[:,'cbgs_id'] = df1.loc[:,'cbgs_id'].apply(drop_last3)
```

```python
#Looks good.
df1
```

```python
#Aggregate the number of visits from the same tract in that week.
df1=df1.groupby(['cbgs_id']).sum()
```

```python
#Reset the index.
df1.reset_index(inplace=True)
```

```python
#Rename the columns.
df1.columns=['tract_id','parkcount_2019']
```

```python
df1
```

```python
df1.to_csv('df1.csv',index=False)
```

```python
link="2020-4-cleaned.csv"

df2=pd.read_csv(link)
```

```python
df2
```

```python
df2['cbgs_id'] = df2['cbgs_id'].astype('string')
```

```python
drop_last1 = lambda s:s[:-1]
df2.loc[:,'cbgs_id'] = df2.loc[:,'cbgs_id'].apply(drop_last1)
```

```python
df2
```

```python
df2=df2.groupby(['cbgs_id']).sum()
```

```python
df2.reset_index(inplace=True)
```

```python
df2.columns=['tract_id','parkcount_2020']
```

```python
df2
```

```python
df2.to_csv('df2.csv',index=False)
```

```python
#Merge the first two datasets with df1 on the left and df2 on the right.
merge1=df1.merge(df2,left_on='tract_id',right_on='tract_id',how='outer') 
```

```python
merge1
```

```python
#Save the merge1.CSV file.
merge1.to_csv('merge1.csv',index=False)
```
