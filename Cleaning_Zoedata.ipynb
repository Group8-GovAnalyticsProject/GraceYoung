```python
#Read a CSV file downloaded from the U.S. Census Bureau website.
import pandas as pd

location='https://raw.githubusercontent.com/g-raceyoung/PUBPOL542/main/'
fileCSV=location+'Median_age_by_sex.csv'
df=pd.read_csv(fileCSV,skiprows=1)
```

```python
#Check the contents of the data frame.
df
```

```python
#Rename the columns for easy cleanup later.
df.columns=['tract_id','GeographicAreaName','1','2','3','4','5','6']
```

```python
#Drop the columns that are not wanted.
df.drop(columns=['2','4','6'],inplace=True)
```

```python
#Rename the columns.
df.columns=['tract_id','GeographicAreaName','MedianAgeTotal','MedianAgeMale','MedianAgeFemale']
```

```python
#Re-check the table and see how it looks now.
df
```

```python
# Since we will be using the 11-digit tract id as the reference unit, the first 9 characters of each entry need to be removed.
drop_first9 = lambda s:s[-11:]
df.loc[:,'tract_id'] = df.loc[:,'tract_id'].apply(drop_first9)
```

```python
#See how it looks now.
df
```

```python
#Looks clean. Save it as a new CSV file.
df.to_csv('MedianAgebySex.csv',index=False)
```
